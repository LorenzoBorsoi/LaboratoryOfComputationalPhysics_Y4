1\. **Hurricanes per Year**

The number of hurricanes in 2005 was 15. The historic average is 6.3. Is this number signficantly different?
- Assume the number of hurricanes is random, i.e. follows the Poisson distribution.
- Assume as statistically significant a probability that has a Z score of 3 or larger with respect a normal distribution.

**Hint**: compute the probability that in a single year are observed 15 or more hurricances.

````python
import numpy as np
from scipy.stats import norm

# Z-test
x=15
mu=6.3
sigma=np.sqrt(mu)
Z=(x-mu)/sigma
print('Z = %.3f' %Z)

# p-value
p=norm.cdf(2*mu-x,loc=mu,scale=sigma)+1-norm.cdf(x,loc=mu,scale=sigma)                   # one-tailed test (because it has no sense to consider the left tail of the distribution - number of hurricane cannot be less than zero)
print('P-value = %.4f %%' %(p*100))
# Yes, observing 15 hurricanes is significally different from the mean of 6.3 hurricanes per year
````

2\. **Pairwise t-test**

In an experiment, a group of 10 individuals agreed to participate in a study of blood pressure changes following exposure to halogen lighting. Resting systolic blood pressure was recorded for each individual. The participants were then exposed to 20 minutes in a room lit only by halogen lamps. A post-exposure systolic blood pressure reading was recorded for each individual. The results are presented in the following data set:

```python
pre = np.array([120, 132, 120, 110, 115, 128, 120, 112, 110, 100])
post = np.array([140, 156, 145, 130, 117, 148, 137, 119, 127, 135])
```

Determine whether the change in blood pressures within our sample was statistically significant.

**Hint:**
in this case, the Student's $t$-test should be performed to compare the two datasets.
Use the following test statistics:

$$T = \frac{\bar{x}_1 - \bar{x}_2}{\sigma \sqrt{\frac{2}{n}}}$$

and 

$$\sigma = \sqrt{\frac{\sigma_1^2 + \sigma_2^2}{2}}$$

````python
from scipy.stats import t

# T-test
pre = np.array([120, 132, 120, 110, 115, 128, 120, 112, 110, 100])
post = np.array([140, 156, 145, 130, 117, 148, 137, 119, 127, 135])
x1, x2 = np.mean(pre), np.mean(post)
var1, var2 = np.var(pre), np.var(post)
sigma=np.sqrt((var1+var2)/2)
T=(x2-x1)/(sigma*np.sqrt(2/len(pre)))
print('T = %.3f' %T)

# p-value
p=1-t.cdf(x2,loc=x1,scale=sigma,df=len(pre)-1)
print('Probability of observing results such as in post = %.2f %%' %(100*p))
# choosing a significance level of 5%, we determine that the change in blood pressure is not statistically significant.
````

3\. **FFT of a simple dataset**

Perform a periodicity analysis on the lynxs-hares population, i.e. determine what is the period of the population of these animals.

````python
from scipy import fftpack
import matplotlib.pyplot as plt

# importing the data
year, hares, lynxes = [], [], []
n=0
with open('populations.txt','r') as file:
    for line in file:
        if n!=0:
            t=line.split('\t')
            year.append(float(t[0]))
            hares.append(float(t[1]))
            lynxes.append(float(t[2]))
        n+=1
year, hares, lynxes = np.array(year), np.array(hares), np.array(lynxes)

# performing a fourier transform
fft_hares, fft_lynxes = fftpack.fft(hares), fftpack.fft(lynxes)
sample_freq=fftpack.fftfreq(len(hares))
power_hares, power_lynxes = np.abs(fft_hares), np.abs(fft_lynxes)

# plotting the frequency spectrum
fig, ax = plt.subplots(1,2,figsize=(14,5))
ax[0].plot(sample_freq,power_hares,'.')
ax[0].set_xlabel('Frequency  [Hz]')
ax[0].set_ylabel('Power')
ax[0].set_title('Hares')
ax[1].plot(sample_freq,power_lynxes,'r.')
ax[1].set_xlabel('Frequency  [Hz]')
ax[1].set_ylabel('Power')
ax[1].set_title('Lynxes')
plt.show()

# finding the frequency of the data trend
sfcopy=sample_freq.copy()
sort_hares, sort_lynxes = sorted(zip(power_hares,sample_freq),reverse=True), sorted(zip(power_lynxes,sfcopy),reverse=True)
power_hares, sample_freq = zip(*sort_hares)
power_lynxes, sfcopy = zip(*sort_lynxes)

T_hares, T_lynxes = 1/sample_freq[1], 1/sfcopy[1]
print('Frequency of hares population: %.3f 1/year' %sample_freq[1])
print('Frequency of lynxes population: %.3f 1/year' %sfcopy[1])
print('Period of hares population: %.1f year' %T_hares)
print('Period of lynxes population: %.1f year' %T_lynxes)

# we ignore the fact that the maximum power is located at f=0Hz. This may be caused by the limited amount of available data,
    # (mostly in the region of the peaks). In fact, the data can seem mostly plain, and this explains the fact that at 0 Hz we
    # observe the main component of the frequency spectrum.
````

4\. **FFT of an image**

Write a filter that removes the periodic noise from the `moonlanding.png` image by using a 2-dimensional FFT.

* Import the image as a 2D numpy array using `plt.imread("moonlanding.png")`. Examine the image with `plt.imshow()`, which is heavily contaminated with periodic noise.
* Check the documentation of the `scipy.fftpack` package, and find the method that performs a 2D FFT. Plot the spectrum (Fourier transform of) the image. **Hint**: use `LogNorm` to plot the colors in log scale:
```Python
from matplotlib.colors import LogNorm
plt.imshow(image, norm=LogNorm(vmin=5))
```
* Inspect the spectrum, and try to locate the regions of the power spectrum that contain the signal and those which contain the periodic noise. Use array slicing to set the noise regions to zero.
* Apply the inverse Fourier transform to plot the resulting image.

````python
from matplotlib.colors import LogNorm

# importing the image
image=plt.imread('moonlanding.png')
fig, ax = plt.subplots(1,2,figsize=(14,5))
ax[0].imshow(image,plt.cm.gray)
ax[0].set_title('Noised image')

# performing a Fourier transform
fft=fftpack.fft2(image)
power=np.abs(fft)

# plotting the frequency spectrum
fig2=plt.figure()
plt.imshow(power,norm=LogNorm(vmin=5))
plt.title('Fourier transform')
plt.colorbar()

# filtering the spectrum and plotting the result
fft[power>1000]=0
filtered_image=np.real(fftpack.ifft2(fft))
ax[1].imshow(filtered_image,plt.cm.gray)
ax[1].set_title('Filtered image')
````
