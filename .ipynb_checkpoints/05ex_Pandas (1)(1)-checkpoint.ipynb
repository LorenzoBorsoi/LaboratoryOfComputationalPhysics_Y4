## Pandas analysis

This exercise consists in analyzing a dataset containg timing information from a series of Time-to-Digital-Converters (TDC) implemented in a couple of FPGAs. Each measurement (i.e. each row of the input file) consists of a flag that specifies the type of message ('HEAD', which in this case is always 1), two addresses of the TDC providing the signal ('FPGA' and 'TDC_CHANNEL'), and the timing information ('ORBIT_CNT', 'BX_COUNTER', and 'TDC_MEAS'). Each TDC count corresponds to 25/30 ns, whereas a unit of BX_COUNTER corresponds to 25 ns, and the ORBIT_CNT is increased every 'x' BX_COUNTER. This allows to store the time in a similar way to hours, minutes and seconds.

1\. Create a Pandas DataFrame reading N rows of the 'data_000637.txt' dataset. Choose N to be smaller than or equal to the maximum number of rows and larger that 10k.

2\. Find out the number of BX in a ORBIT (the value 'x').

3\. Find out how much the data taking lasted. You can either make an estimate based on the fraction of the measurements (rows) you read, or perform this check precisely by reading out the whole dataset.

4\. Create a new column with the absolute time in ns (as a combination of the other three columns with timing information).

5\. Replace the values (all 1) of the HEAD column randomly with 0 or 1.

6\. Create a new DataFrame that contains only the rows with HEAD=1.

7\. Make two occupancy plots (one for each FPGA), i.e. plot the number of counts per TDC channel

8\. Use the groupby method to find out the noisy channels, i.e. the TDC channels with most counts (say the top 3)

9\. Count the number of unique orbits. Count the number of unique orbits with at least one measurement from TDC_CHANNEL=139

```python
import pandas as pd

#question 1 complete
df = pd.read_csv('~/data/data_000637.txt')
df = df[:100000]
orbmin = df['ORBIT_CNT'].min()

x = df['BX_COUNTER'].max()+1
#/(df['ORBIT_CNT'].max() - df['ORBIT_CNT'].min())



#this is the number of times the BX_counter was reset to 0 before continuing to count
BXcycles = df['BX_COUNTER'].value_counts()[x-1]

#question 2 complete
print('The number of BX counts in one orbit is',x)

duration = BXcycles*x

#question 3 complete
print("There was a total of",duration,"bx_counts in my subset of the data, and this corrosponds to",duration*25,"nano seconds.")

#question 4 complete 
def nano(entry):
    orb = entry['ORBIT_CNT']
    bx = entry['BX_COUNTER']
    td = entry['TDC_MEAS']
    
    #nano seconds in the orbit
    norb = orb*x*25
    
    #nano seconds in the bx
    nbx = bx * 25
    
    #nano seconds in the TDC_meas
    ntd = td * (25/30)
    
    return norb+nbx+ntd

df['Nanoseconds_Since_1st_Orbit'] = df.apply(lambda entry: nano(entry), axis = 1)


df.describe()
```

```python
#question 5 complete
import random
import numpy as np
val = np.random.randn(len(df))
val = (np.sign(val)+1)/2

df['HEAD'] = val 
print(val)

#question6 complete
newdf = df[df['HEAD'] == 1]

newdf
```

```python
#question 7 complete 
#Make two occupancy plots (one for each FPGA), i.e. plot the number of counts per TDC channel
fpga1 = df[df['FPGA']==1]
fpga0 = df[df['FPGA']==0]

binss = fpga1['TDC_CHANNEL'].max() - fpga1['TDC_CHANNEL'].min() 

f1h = fpga1.hist(column = 'TDC_CHANNEL',bins = binss)
print(type(f1h))
#,xlabel = 'TDC channel', ylabel = 'Counts')
fpga0.hist(column = 'TDC_CHANNEL',bins = binss)
#title = 'FPGA 2',xlabel = 'TDC channel', ylabel = 'Counts')

print('Plots showing the number of times each TDC channel is used.\n With the TDC channel on the x axis and the count for the number of times it is used on the y axis.')
```

```python
#question 8 complete
tdc = fpga1.groupby(['TDC_CHANNEL'])['TDC_CHANNEL'].count()
tdc = tdc.nlargest(n=3)
print('The TDC channels with the most counts for fpga 1 are\n',tdc)

tdc = fpga0.groupby(['TDC_CHANNEL'])['TDC_CHANNEL'].count()
tdc = tdc.nlargest(n=3)
print('\n\nThe TDC channels with the most counts for fpga 0 are\n',tdc)
```

```python
#Complete
#question 9 Count the number of unique orbits with at least one measurement from TDC_CHANNEL=139

ch139 = df[df['TDC_CHANNEL']==139]

uniorb = len(pd.unique(ch139['ORBIT_CNT']))

print('The number of unique orbits measured by channel 139 is',uniorb)          
```
