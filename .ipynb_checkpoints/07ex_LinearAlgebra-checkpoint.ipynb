1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

# 1st task
seed=np.random.seed(18)
n=540
x1=np.random.normal(size=n).reshape((n,1))
x2=x1+np.random.normal(scale=3, size=n).reshape((n,1))
x3=2*x1+x2
X=np.hstack((x1,x2,x3))

# 2nd task
cov=np.cov(X.T)
e_val, e_vec = la.eig(cov)
e_val=np.real_if_close(e_val)
e_vec=e_vec.T
zipped=sorted(zip(e_val,e_vec), reverse=True)             # reordering both eigenvalues and eigenvector on a descending scale of the eigenvalues
e_val, e_vec = zip(*zipped)
e_val, e_vec = np.array(e_val), np.array(e_vec).T

# 3rd task
U, spectrum, Vt = la.svd(cov)
print('Same eigenvalues?  ',np.allclose(e_val,spectrum))
print('Same eigenvectors?  ',np.allclose(e_vec,U))

# 4th and 5th tasks
print('\nVariability explained from 1st principal component:\t\t%.1f%%' % (100*e_val[0]/np.sum(e_val)) )
print('Variability explained from 1st and 2nd principal components:\t%.1f%%' % (100*(e_val[0]+e_val[1])/np.sum(e_val)) )
print('Variability explained from all principal components:\t\t%.1f%%' % (100*(e_val[0]+e_val[1]+e_val[2])/np.sum(e_val)) )
# This results makes sense. The 1st principal component exlpain most of the variation (about 93%), while the second component
    # is much less significative. This is due to the fact that the variable x2 is just x1 with some random noise
    # x3 was obtaines in a deterministic way from x1 and x2, so that component doesn't introduce any variation in the dataset
X=X.T
Xr=np.dot(e_vec.T,X)                                       # change of basis
Xrr=Xr[[0,1],:]                                            # reducing the data taking only the 1st and 2nd features (the most significant)
va, ve=la.eig(np.cov(Xrr))
va=np.real_if_close(va)
plt.scatter(Xrr[0,:], Xrr[1,:], alpha=0.2)                 # plotting a graph to show the change of basis
for li, vi in zip(va, np.diag([1]*2)):
    plt.plot([0, li*vi[0]], [0, li*vi[1]], 'r-', lw=2)
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.title('plot showing x2 vs x1 in the eigenvector basis')
plt.show()

# 6th task
figure, ax = plt.subplots(2,3,figsize=(18,12))
ax[0][0].scatter(X[0,:],X[1,:],alpha=0.2)
ax[0][0].set_title('x2 vs x1')
ax[0][1].scatter(X[0,:],X[2,:],alpha=0.2)
ax[0][1].set_title('x3 vs x1')
ax[0][2].scatter(X[1,:],X[2,:],alpha=0.2)
ax[0][2].set_title('x3 vs x2')
ax[1][0].scatter(Xr[0,:],Xr[1,:],alpha=0.2)
ax[1][0].set_title('x2 vs x1 - eigenvector basis')
ax[1][1].scatter(Xr[0,:],Xr[2,:],alpha=0.2)
ax[1][1].set_title('x3 vs x1 - eigenvector basis')
ax[1][2].scatter(Xr[1,:],Xr[2,:],alpha=0.2)
ax[1][2].set_title('x3 vs x2 - eigenvector basis')
```

2\. **PCA on a nD dataset**

Start from the dataset you have genereted in the previous exercise and add uncorrelated random noise. Such noise should be represented by other 10 uncorrelated variables normal distributed, with standar deviation much smaller (say, a factor 50) than those used to generate the $x_1$ and $x_2$.

Repeat the PCA procedure and compare the results with what you obtained before

```python
noise=np.random.normal(scale=0.05,size=(10,n))
Xn=np.vstack((X,noise))

e_val, e_vec = la.eig(np.cov(Xn))
e_val=np.real_if_close(e_val)
e_vec=e_vec.T
zipped=sorted(zip(e_val,e_vec),reverse=True)           # same as in previous exercise, ordering eigenvalues and eigenvectors
e_val, e_vec = zip(*zipped)
e_val, e_vec = np.array(e_val), np.array(e_vec).T

variabilities=[100*sum(e_val[0:i+1])/sum(e_val) for i in range(len(e_val))]       # calculating the percentages of variation
for i in range(len(variabilities)):
    print('Variability explaines by first %i components:\t%.3f %%' %(i+1,variabilities[i]))
    
n_for_99=0              # number of feature to consider for having a *soglia* % of variability explained by the data
soglia=99               # threshold for the percentage of variability
for i in range(len(variabilities)):     # checking for n_for_99
    n_for_99+=1
    if variabilities[i]>soglia:
        break
        
Xnr=np.dot(e_vec.T,Xn)       # change of basis
Xnrr=Xnr[0:n_for_99,:]       # reducing the data
print('\nShape of the reduced data:\t',Xnrr.shape)
```

```python
import csv
import pandas as pd
import seaborn as sns

# importing the data
X, y = [], []
with open('magic04.data') as data_file:
    for line in csv.reader(data_file,delimiter=','):
        X.append(list(map(float,line[:-1])))
        y.append(line[-1])
X, y = np.array(X).T, np.array(y)
print('Shape of original data (excluding "categories"):',X.shape,'\n')

# finding eigenvalues and eigenvectors, ordering them
e_val, e_vec = la.eig(np.cov(X))
e_val, e_vec = np.real_if_close(e_val), e_vec.T
zipped=sorted(zip(e_val, e_vec),reverse=True)
e_val, e_vec = zip(*zipped)
e_val, e_vec = np.array(e_val), np.array(e_vec).T

# find number of features to be considered for explain a threshold of variability
variabilities=[100*sum(e_val[0:i+1])/sum(e_val) for i in range(len(e_val))]
for i in range(len(variabilities)):
    print('Variability explaines by first %i components:\t%.3f %%' %(i+1,variabilities[i]))
n_for_99=0
soglia=99
for i in range(len(variabilities)):
    n_for_99+=1
    if variabilities[i]>soglia:
        break

# change basis for data representation, reducing data
Xr=np.dot(e_vec.T,X)
Xrr=Xr[0:n_for_99,:]
print('\nShape of the reduced data:\t',Xrr.shape)

# plotting the two most significant features, with different colors for different type of observation of the detectors
Xrr, y = np.array(Xrr,dtype=object), np.array(y,dtype=object)
data=pd.DataFrame(np.random.permutation(np.hstack((Xrr.T,y.reshape(19020,1)))))
g=sns.relplot(x=0,y=1,hue=6,data=data.loc[:1000,:]).set(title='First two most significant features, both in the eigenvector basis')
```
