1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python
import pandas as pd
import numpy as np
from scipy import linalg as la
import matplotlib.pyplot as plt
import numpy.random as npr
npr.seed(9324988)
%matplotlib inline

N = 3000
x1 = npr.normal(loc = 0, scale = 1, size = N)
x2 = x1 + npr.normal(loc = 0, scale = 3, size = N)
x3 = 2*x1 + x2

dataset = np.zeros((3,N))
dataset[0] = x1
dataset[1] = x2
dataset[2] = x3

covm = np.cov(dataset)
print("Covariance matrix: \n{}".format(covm))

print("\nEig method:")
l, V = la.eig(covm)
l = np.real_if_close(l)
for li,vi in zip(l,V.T):
    print(round(li,2),vi)

print("\nSVD method:")
U,D,VT = la.svd(dataset)
for di,ui in zip(D**2/(N-1),U.T):
    print(round(di,2),ui)
    
spect = D**2/(N-1)
tr = np.sum(spect)
print("\n")
for eig in spect:
    print("The eigenvalue {} influences the data for the {}%.".format(round(eig,2),round(eig*100/tr,2)))
    
print("As we can see there is useless data, this is to be expected since we built the third row of the dataset completely dependent from the other 2.")
print("Thus we can safely remove it without a major loss of information (more than 99% is retained).")

dataset1 = dataset[:2,:]
Xp = np.dot(U.T, dataset)

fig, graph = plt.subplots(nrows = 2, ncols = 3, figsize = (20,10))
graph[0,0].scatter(dataset[0],dataset[1],alpha = 0.4, color = "purple")
graph[0,1].scatter(dataset[0],dataset[2],alpha = 0.4, color = "purple")
graph[0,2].scatter(dataset[1],dataset[2],alpha = 0.4, color = "purple")
graph[1,0].scatter(Xp[0],Xp[1],alpha = 0.4, color = "green")
graph[1,1].scatter(Xp[0],Xp[2],alpha = 0.4, color = "green")
graph[1,2].scatter(Xp[1],Xp[2],alpha = 0.4, color = "green")
plt.show()

print("\nApplying PCA:\n")
Xp = Xp[:2,:]
U,D,VT = la.svd(Xp)
spect = D**2/(N-1)
tr = np.sum(spect)
for eig in spect:
    print("The eigenvalue {} influences the data for the {}%.".format(round(eig,2),round(eig*100/tr,2)))
```

2\. **PCA on a nD dataset**

Start from the dataset you have genereted in the previous exercise and add uncorrelated random noise. Such noise should be represented by other 10 uncorrelated variables normal distributed, with standar deviation much smaller (say, a factor 50) than those used to generate the $x_1$ and $x_2$.

Repeat the PCA procedure and compare the results with what you obtained before

```python
data = np.zeros((13,N))
data[0:3,:] = dataset
for i in range(3,13):
    data[i] = npr.normal(loc = 0, scale = 1/50, size = N)

U,D,VT = la.svd(data)
for di,ui in zip(D**2/(N-1),U.T):
    print(round(di,6),ui)
covm = np.cov(data)    
spect = D**2/(N-1)
tr = np.sum(spect)
print("\n")
for eig in spect:
    print("The eigenvalue {} influences the data for the {}%.".format(round(eig,6),round(eig*100/tr,6)))

plt.scatter(data[0],data[4],alpha = 0.4, color = "purple")
plt.ylim((-5,5))
plt.title("Scatter plot of fouth feature versus first feature")
plt.show()
print("\n\nAPPLYING PCA:")
datap = np.dot(U.T, data)
datap = datap[:2,:]
plt.scatter(datap[0],datap[1],alpha = 0.4, color = "purple")
plt.title("Scatter plot of second feature versus first feature")
plt.show()
```

3 \. **Looking at an oscillating spring** (optional)

Imagine you have $n$ cameras looking at a spring oscillating along the $x$ axis. Each  camera record the motion of the spring looking at it along a given direction defined by the pair $(\theta_i, \phi_i)$, the angles in spherical coordinates. 

Start from the simulation of the records (say ${\cal O}(1000)$) of the spring's motion along the x axis, assuming a little random noise affects the measurements along the $y$. Rotate such dataset to emulate the records of each camera.

Perform a Principal Component Analysis on the thus obtained dataset, aiming at finding the only one coordinate that really matters.


```python
def rot(dat,th,ph):
    Ry = np.array([[np.cos(th),0,-np.sin(th)],[0,1,0],[np.sin(th),0,np.cos(th)]])
    Rz = np.array([[np.cos(ph),0,np.sin(ph)],[-np.sin(ph),0,np.cos(ph)],[0,0,1]])
    return np.dot(Rz,np.dot(Ry,dat))

N = 1000
a = 5
k = 4*np.pi
time = np.linspace(0,2000, num = N)
x = a * np.sin(k * time)
y = 2 * npr.random(N) -1
plt.plot(time, x, ls = "", marker = "+")
plt.show()
plt.plot(x, y, ls = "", marker = "+")
plt.show()

data = np.zeros((3,1000), dtype="float")
data[0] = x
data[1] = y

print("\n")
U,D,VT = la.svd(data)
for di,ui in zip(D**2/(N-1),U.T):
    print(round(di,2),ui)
covm = np.cov(data)  
spect = D**2/(N-1)
tr = np.sum(spect)
print("\n")
for eig in spect:
    print("The eigenvalue {} influences the data for the {}%.".format(round(eig,6),round(eig*100/tr,6)))

n = 10
for i in range(n):
    theta = np.pi*npr.random()
    phi = 2*np.pi*npr.random()
    dat1 = rot(data, theta, phi)
    #plt.plot(dat1[0], dat1[1], ls = "", marker = "+")
    #plt.xlim([-5,5])
    #plt.ylim([-1,1])
    #plt.show()

print("\n")
U,D,VT = la.svd(dat1)
for di,ui in zip(D**2/(N-1),U.T):
    print(round(di,2),ui)
covm = np.cov(dat1)  
spect = D**2/(N-1)
tr = np.sum(spect)
print("\n")
for eig in spect:
    print("The eigenvalue {} influences the data for the {}%.".format(round(eig,6),round(eig*100/tr,6)))
    
print("\nOnly the data on the x axis is significant.")
```
